```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "name": ["Alice", "BOB", "Charlie", "John"],
    "Age": [25, 30, 35, np.nan]
}

df = pd.DataFrame(data)
age_mean = df["Age"].mean()
df["Age"] = df["Age"].fillna(age_mean)
plt.bar(df["name"], df["Age"])
plt.show()
```
