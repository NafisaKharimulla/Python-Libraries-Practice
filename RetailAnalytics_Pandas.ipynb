```python
import pandas as pd
```

```python
data = {
    "order_id": [1, 2, 3],
    "customer_id": [101, 102, 103],
    "order_date": ["2024-01-10", "2024-02-15", "2024-03-20"],
    "order_amount": [250, 400, 300]
}

df = pd.DataFrame(data)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df["month"] = df["order_date"].dt.month
```

```python
monthly_revenue = df.groupby("month")["order_amount"].sum()
print("Monthly Total Revenue:")
print(monthly_revenue)
```

```python
unique_customers = df.groupby("month")["customer_id"].nunique()
highest_customer_month = unique_customers.idxmax()

print("\nMonth with highest unique customers:", highest_customer_month)
```
