```python
import numpy as np

# Create ratings for 100 employees and 4 quarters
ratings = np.random.randint(1, 6, (100, 4))

# Min-Max Normalization
min_val = ratings.min()
max_val = ratings.max()

normalized = (ratings - min_val) / (max_val - min_val)

#  Average rating per employee
employee_avg = ratings.mean(axis=1)

# Company average rating
company_avg = employee_avg.mean()

# Employees with above-average rating
above_avg_employees = np.where(employee_avg > company_avg)

# Output
print("Normalized ratings (first 3 employees):")
print(normalized[:3])

print("\nAverage rating per employee (first 3):")
print(employee_avg[:3])

print("\nCompany average rating:", company_avg)

print("\nEmployees above company average:")
print(above_avg_employees)
```
