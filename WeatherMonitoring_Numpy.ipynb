```python
import numpy as np
temp_readings=np.random.randint(15,35, (7,24))
print(temp_readings)
```

```python
Daily_max=temp_readings.max(axis=1)
Daily_min=temp_readings.min(axis=1)
print(Daily_max)
print(Daily_min)
```

```python
# Step 3: Day with the largest temperature variation
daily_variation=Daily_max-Daily_min
largest_variation=np.argmax(daily_variation)
print(largest_variation)
```

```python
# Step 4: Replace outliers beyond Â±2 standard deviations with mean temperature
mean_temp = temp_readings.mean()
std_temp = temp_readings.std()

# Find outliers
outliers = (temp_readings > mean_temp + 2*std_temp) | (temp_readings < mean_temp - 2*std_temp)

# Replace outliers with mean
temp_readings[outliers] = mean_temp

print("\nTemperature array after replacing outliers:")
print(temp_readings)
```
