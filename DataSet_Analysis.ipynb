```python
plt.figure(figsize=(10, 5))

plt.plot(df["Date"], df["Close"])
plt.xticks(df["Date"][::400], rotation=45)  # show fewer dates

plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Silver Closing Price Over Time")

plt.show()

```

```python
plt.figure(figsize=(10, 5))

plt.plot(df["Date"] , df["High"], label="High")
plt.plot(df["Date"], df["Low"], label="Low")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("High vs Low Silver Prices")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.hist(df["Close"], bins=20)
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.title("Distribution of Silver Closing Prices")

plt.show()
```

```python
plt.figure(figsize=(10, 4))

plt.plot(df["Date"], df["Volume"])
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Silver Trading Volume Over Time")

plt.show()
```

```python
df["MA_30"] = df["Close"].rolling(30).mean()

plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Close"], label="Close")
plt.plot(df["Date"], df["MA_30"], label="30-Day Moving Avg")

plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Silver Price with Moving Average")
plt.legend()

plt.show()
```
