```python
Data={
    "name":["Alice","BoB","Charlie","Maxy"],
    "Age":[20,30,np.nan,50]
}

df=pd.DataFrame(Data)
avg=df.Age.mean()
df["Age"]=df["Age"].fillna(df["Age"].mean())

plt.bar(df["name"],df["Age"])
plt.show()
```

```python
Data={
    "Dept" :["HR","Finance","HR","IT","HR","IT"],
    "Sal" : [500,100,200,300,400,500]
}
df=pd.DataFrame(Data)
print(df)
Sal_By_Dept=df.groupby("Dept")["Sal"].mean()
print(Sal_By_Dept)

plt.bar(Sal_By_Dept.index, Sal_By_Dept.values)
plt.show()
```

```python
Data={
    "hours_studied":[5,10,8,9,11],
    "score":[80,70,60,30,50]
}
df=pd.DataFrame(Data)
plt.scatter(df["hours_studied"],df["score"])
corr = df["hours_studied"].corr(df["score"])
print(corr)
```

```python
data={
    "gender":["male","female","male","female","male"]
}
df=pd.DataFrame(data)
counts=df["gender"].value_counts()
plt.pie(counts,labels=counts.index,autopct="%1.1f%%")
plt.show()
```
