```python
import numpy as np
# Sample data (30 days, 5 products)
np.random.seed(0)
sales = np.random.randint(-20, 100, size=(30, 5))

# Replace negative values with 0
sales[sales < 0] = 0

#Compute weekly averages
# Only first 28 days are used (4 full weeks)
weekly_sales = sales[:28].reshape(4, 7, 5)
avg_weekly_sales = weekly_sales.mean(axis=1)

# Average weekly sales per product
avg_per_product = avg_weekly_sales.mean(axis=0)

#  Product with highest average weekly sales
best_product = np.argmax(avg_per_product)

print("Average weekly sales per product:")
print(avg_per_product)

print("\nProduct with highest average weekly sales:", best_product)
```
